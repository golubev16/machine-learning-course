```python
import numpy as np
import pandas as pd
import random
import warnings
warnings.filterwarnings("ignore")
from category_encoders import TargetEncoder
```

```python
data = pd.read_csv("Train.csv")
data
```

```python
a = []
for i in range(1,24):
    a.append(i)
X = data.iloc[:,a]
y = data.iloc [:, 24]
```

```python
build_t = []
balcons = []
g_lift = []
years = []
months = []
metro_dist = []
area = []
for index, row in X.iterrows():
    area.append(row['area'])
    a = row['date'].split('-')
    years.append(int(a[0]))
    months.append(int(a[1]))
    if np.isnan(row['build_tech']):
        r = random.randint(0, 71)
        build_t.append(0.0 if r <= 36 else 1.0 if r <= 67 else 2.0)
    else:
        build_t.append(row['build_tech'])
    if np.isnan(row['g_lift']):
        g_lift.append(0.0 if random.randint(0,1) == 0 else 1.0)
    else:
        g_lift.append(row['g_lift'])
    if np.isnan(row['metro_dist']):
        metro_dist.append(X['metro_dist'].mean())
    else:
        metro_dist.append(row['metro_dist'])
X.drop('metro_dist', axis=1, inplace=True)
X.drop('g_lift', axis=1, inplace=True)
X.drop('build_tech', axis=1, inplace=True)
X.drop('date', axis=1, inplace=True)
X.drop('balcon', axis=1, inplace=True)
X['g_lift'] = g_lift
X['build_tech'] = build_t
X['year'] = years
X['month'] = months
X['metro_dist'] = metro_dist
X
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (X, y, test_size = 0.25, random_state = 0)
```

```python
feat_te = ['street_id', 'build_tech', 'area']
feat1 = ['street_id_t', 'build_tech_t', 'area_t']
```

```python
te = TargetEncoder(cols = feat_te)
te.fit(x_train[feat_te], y_train)
```

```python
x_train_TE = x_train.join(te.transform(x_train[feat_te]).add_suffix('_T'))
x_test_TE = x_test.join(te.transform(x_test[feat_te]).add_suffix('_T'))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform (x_train_TE)
x_test = sc.transform (x_test_TE)
```

```python
import lightgbm as lgb
d_train = lgb.Dataset (x_train, label = y_train)
params = {}
params ['learning_rate'] = 0.36
params ['metric'] = 'mae'
params ['n_estimators'] = 379
params ['force_col_wise'] = True

clf = lgb.train (params, train_set = d_train)
```

```python
y_pred = clf.predict (x_test)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
