```python
from tqdm import tqdm
import numpy as np
import pandas as pd
import folium
import matplotlib as mpl
import matplotlib.cm as cm
pd.options.display.max_columns = 200
```

```python
df = pd.read_csv("Rus_schools_final.csv", encoding='1251')
df.head(3)
```

```python
center = {'lat': 54.621605, 'lon': 39.708499}
```

```python
map_folium = folium.Map(location=(center['lat'], center['lon']),
                        tiles='Stamen Toner',
                        zoom_start=11)
```

```python
for ind, row in df.iterrows():
    if abs(row['lat'] - center['lat']) < 0.2 and abs(row['lon'] - center['lon']) < 0.2:
        if df.values[ind][1] == 'Муниципальное образовательное учреждение "Средняя общеобразовательная школа № 35 , г. Рязани"':
            folium.CircleMarker([row['lat'], row['lon']],
                            radius=15,
                            fill=True, 
                            fill_opacity=0.9,
                            color = 'red'
                           ).add_to(map_folium)
            continue
        folium.CircleMarker([row['lat'], row['lon']],
                            radius=7,
                            fill=True, 
                            fill_opacity=0.9
                           ).add_to(map_folium)
map_folium
```
